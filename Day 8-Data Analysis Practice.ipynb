```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('train.csv')
```

```python
df
```

```python
df['Family Size']=df['SibSp']+df['Parch']+1
```

```python
df.head(5)
```

```python
def type(num):
    if num==1:
        return 'Alone'
    elif num >1 and num <5:
        return 'Small'
    else:
        return 'Larger'
```

```python
df['Family_Type']=df['Family Size'].apply(type)
```

```python
df['Cabin']=df['Cabin'].fillna('M')
```
